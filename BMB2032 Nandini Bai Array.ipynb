```python
#Elementwise arithmetic operation
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```

```python
# Matrics multiplication
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print(arr1@arr2)
```

```python
# Concatenate
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
concatenate_arr=np.concatenate((arr1,arr2),axis=0)
print(concatenate_arr)
print()
concatenate_arr=np.concatenate((arr1,arr2),axis=1)
print(concatenate_arr)
```

```python
# Transpose : it converts coloum into row and row into coloum (3*2--->3*2)
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print()
transpose_arr=arr.transpose()
print(transpose_arr)
```

```python
# Flattened : it will going to convert the multidimension into single dimension 2D to 1D
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print()
flattened_arr=arr.flatten()
print(flattened_arr)
```

```python
# To check real no and complex & scalar no
import numpy as np
a=np.array([1+1j,1+0j,4.5,3,2,2j])
print("Original array")
print(a)
print("Checking for complex number:")
print(np.iscomplex(a))
print("Checking for real number:")
print(np.isreal(a))
print("Checking for scalar number:")
print(np.isscalar(3.1))
print(np.isscalar([3.1]))
```
