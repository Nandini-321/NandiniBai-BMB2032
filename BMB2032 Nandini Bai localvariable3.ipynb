```python
#scope of local variable
def fun(x):
  print("local x is:",x)
  x=2
  print("changed local x to:",x)
x=50
fun(x)
print("x is still:",x)
```

```python
# Global variables: r the key words used to
def fun1():
  global y
  print("Y is:",y)
  y=2
  print("changed y to:",y)
y=50
fun1()
print("value of y is:",y)
```

```python
def fun():
  y=2
  print("changed y to:",y)
y=50
fun()
print("y is still:",y)
```
