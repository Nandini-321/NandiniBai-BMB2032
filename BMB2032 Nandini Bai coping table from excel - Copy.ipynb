```python
# To filter the data(only wheat)
wheat_data=df[df["Crop"]=="Wheat"]
print("\nwheat Data:")
print(wheat_data)
```

```python
# case sensitive
wheat_data=df[df["Crop"]=="wheat"]
print("\nwheat Data:")
print(wheat_data)
```

```python
# Group by year and sum production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nYearly Production:")
print(yearly_production)
```

```python
# Scatter plot (rainfall vs production)
import matplotlib.pyplot as plt

plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall(mm)")
plt.ylabel("Production(Tons)")
plt.title("Rainfall vs Production")
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8,5))
plt.bar(yearly_production.index,yearly_production.values,color="skyblue")
plt.xlabel("Year")
plt.ylabel("Total production(Tons)")
plt.title("total Crop Production per Year")
plt.xticks(yearly_production.index)
plt.show()
```

```python
# Group by year and sum production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print(yearly_production)
print(yearly_production.index)
print(yearly_production.values)
```

```python
# Bar chart area by year
area_by_year=df.groupby("Year")["Area_Hectares"].sum()
area_by_year.plot(kind="bar")
plt.xlabel("Year")
plt.ylabel("Total Area(Hectares)")
plt.title("Total Area by Year")
plt.show()
```

```python
#Line plot: Production over years without for loop
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]

plt.figure()

plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
```

```python
wheat=df[df["Crop"]=="Wheat"]
rice=df[df["Crop"]=="Rice"]

plt.figure()

plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")
plt.xlabel("Year")
plt.ylabel("Production(Tons)")
plt.title("Crop Production Over Years")
plt.legend()
plt.show()
```

```python
data_2022=df[df["Year"]==2022]
print(data_2022)

plt.figure(figsize=(6,4))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=['green','orange'])

plt.xlabel("Crop")
plt.ylabel("Production(Tons)")
plt.title("Production by Crop in 2022")

plt.show()

```

```python
# custom styled bar plot
plt.figure(figsize=(8,5))

bars=plt.bar(yearly_production.index,
             yearly_production.values,
             color='green',
             edgecolor='black')
for bar in bars:
  height=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,
    height,
    str(height),
    ha='center',
    va='bottom')
```
