```python
import random

inventory = {}
total_money = 0

fish_price = {
    "Rahu": 12,
    "Catla": 18,
    "Sinchu": 25
}

print("\nFishing Game Menu")
print("1 Catch Fish")
print("2 View Inventory")
print("3 Sell Fish")
print("4 Exit")

while True:
    try:
        choice = int(input("Choose an option: "))
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

    if choice == 1:
        fish = random.choice(list(fish_price.keys()))
        inventory[fish] = inventory.get(fish, 0) + 1
        print(f"You caught a {fish}!")
    elif choice == 2:
        if not inventory:
            print("No fish in your inventory.")
        else:
            print("Your inventory:")
            for fish, count in inventory.items():
                print(f"{fish}: {count}")
    elif choice == 3:
        if not inventory:
            print("No fish to sell.")
        else:
            earnings = 0
            for fish, count in inventory.items():
                earnings += fish_price[fish] * count
            total_money += earnings
            inventory.clear()
            print(f"You sold all fish for ${earnings}!")
            print(f"Total money: ${total_money}")
    elif choice == 4:
        print("Exiting Fishing Game. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
# with parameter with return
def fact(m):
  fact=1
  if (m>0):
    for i in range (1,m+1):
        fact=fact*i
        return(fact)

  else:
   print("enter a valid number")

n=int(input("enter a number"))
z=fact(n)
print(z)
```
