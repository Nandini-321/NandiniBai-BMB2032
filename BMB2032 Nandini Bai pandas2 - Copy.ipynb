```python
import pandas as pd
import numpy as np
# Create a data with missing values
data = pd.DataFrame({
    'Plot': ['A', 'B', 'C', 'D', 'E'],
    'yield_kg': [2500, np.nan, 1800, 1700, 1600],
    'rainfall': [150, 200, np.nan, 180, 170],
    'temperature': [28, 30, 29, np.nan, 27]
})

print("=== Data with Missing Values ===")
print(data)

# === DETECTING MISSING DATA ===
print("\n=== Detecting Missing Values ===")
print("Is null?")
print(data.isnull())

print("\nTotal missing per column:")
print(data.isnull().sum())
```

```python
# Filling the missing values
print("\n===filling stratergies===")
# fill with mean
data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean)
print("\nfill with mean:")
print(data_mean['yield_kg'])
```

```python
# Fill with median
data_median=data.copy()
print(data_median['yield_kg'])
data_median['yield_kg']=data_median['yield_kg'].fillna(data_median['yield_kg'].median)
print("\nfill with median:")
print(data_median['yield_kg'])
```

```python
# Forward fill (use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print('\nforward fill')
print(data_ffill)
```

```python
# Dropping the missing data
print("\n>>>droppong roows<<<")
data_dropped=data.dropna()
print(data_dropped)
print(f"original rows:{len(data)},after drop:{len(data_dropped)}")
```
