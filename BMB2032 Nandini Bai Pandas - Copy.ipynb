```python
# we can create own indexes
import pandas as pd
production=pd.Series([30,60,45,25],
                     index=["Rice","Wheat","Maize","Barley"])
print(production)
```

```python
#loc(location by label)
print(production["Rice"])
print(production.loc["Rice"])
```

```python
# Max ,Min , mean, sum,(idxmax(print the index label of max value))
print(production.sum())
print(production.max())
print(production.mean())
print(production.idxmax())
```

```python
print(production.sum())
print(production.max())
print(production.mean())
print(production.idxmax())
print(production.idxmax(),"is having maximum value of ",production.max())
```

```python
# Increase all production by 10%
increase_production=production*1.10
print(increase_production)
```

```python
# Production greater than 40%
high_production=production[production>40]
print(high_production)
print()
print(production>40)
```

```python
import pandas as pd
data={
    "farm":["farm A","farm B","farm C","farm D"],
    "crop":["wheat","rice","wheat","maize"],
    "area_hectare":[10,15,8,12],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
# for the existing one want to add a new column to it
df["yield_tons_per_hectare"]=df["production_tons"]/df["area_hectare"]
print(df)
```

```python
# Farms greater than 3
# Fixed typo: changed 'yield_ton_per_hectare' to 'yield_tons_per_hectare'
high_yield=df[df["yield_tons_per_hectare"]>3]
print(high_yield)
```

```python
# sort the farms by highest production
sorted_df=df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```
